```python
img_t=cv2.imread('/content/th (1).jpg',cv2.IMREAD_GRAYSCALE)
plt.imshow(img_t,cmap='gray')
plt.show()
```

```python
_, binary_image =cv2.threshold(img_t,127,255,cv2.THRESH_BINARY)
plt.imshow(binary_image,cmap='gray')
plt.show()
```

```python
kernal =np.ones((5,5),np.uint8)
dilation=cv2.dilate(binary_image,kernal,iterations=1)
plt.imshow(dilation,cmap='gray')
plt.show()
```

```python
tx, ty = 50, 100
M = np.float32([[1, 0, tx], [0, 1, ty]])
translated_img = cv2.warpAffine(img, M, (img.shape[1], img.shape[0]))
translated_img_rgb = cv2.cvtColor(translated_img, cv2.COLOR_BGR2RGB)
plt.imshow(translated_img_rgb)
plt.axis('off')
plt.title('Translated image')
plt.show()
```

```python
rotation_matrix = cv2.getRotationMatrix2D((img_rgb.shape[1] / 2, img_rgb.shape[0] / 2), -45, 1)
rotated = cv2.warpAffine(img_rgb, rotation_matrix, (img_rgb.shape[1], img_rgb.shape[0]))
rotated_rgb = cv2.cvtColor(rotated, cv2.COLOR_BGR2RGB)
plt.imshow(rotated_rgb)
plt.axis('off')


img = cv2.imread('th (1).jpg')
flip_code = 1
reflected_img_rgb = cv2.flip(rotated_rgb, flip_code)
reflected_img_rgb = cv2.cvtColor(reflected_img_rgb, cv2.COLOR_BGR2RGB)

plt.imshow(reflected_img_rgb)
plt.axis('off')
plt.title('Rotated and Reflected image')
plt.show()
```

```python
img = cv2.imread('th (1).jpg')
shear_factor_x = 0.5
shear_factor_y = 0.0
M = np.float32([[1, shear_factor_x, 0], [shear_factor_y, 1, 0]])

sheared_img = cv2.warpAffine(img, M, (img.shape[1], img.shape[0]))
sheared_img_rgb = cv2.cvtColor(sheared_img, cv2.COLOR_BGR2RGB)
plt.imshow(sheared_img_rgb)
plt.axis('off')

rotation_matrix = cv2.getRotationMatrix2D((sheared_img_rgb.shape[1] / 2, sheared_img_rgb.shape[0] / 2), -45, 1)
rotated = cv2.warpAffine(sheared_img_rgb, rotation_matrix, (sheared_img_rgb.shape[1], sheared_img_rgb.shape[0]))
rotated_rgb = cv2.cvtColor(rotated, cv2.COLOR_BGR2RGB)
plt.imshow(rotated_rgb)
plt.axis('off')
plt.title('Rotated and Sheared image')
plt.show()
```
